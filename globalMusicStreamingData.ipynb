```python
import pandas as pd

import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Streaming Platform', palette='viridis')
plt.title('Platform Distribution')
plt.xlabel('Streaming Platform')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Top Genre', order=df['Top Genre'].value_counts().index, palette='plasma')
plt.title('Top Genre Distribution')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Minutes Streamed Per Day', hue='Streaming Platform', palette='viridis')
plt.title('Age vs. Minutes Streamed Per Day')
plt.xlabel('Age')
plt.ylabel('Minutes Streamed')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Subscription Type', y='Repeat Song Rate (%)', palette='coolwarm')
plt.title('Subscription Type vs. Repeat Song Rate')
plt.xlabel('Subscription Type')
plt.ylabel('Repeat Song Rate (%)')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
top_artists = df['Most Played Artist'].value_counts().head(5)
sns.barplot(x=top_artists.index, y=top_artists.values, palette='magma')
plt.title('Top 5 Most Played Artists')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.pairplot(df, hue='Streaming Platform', palette='husl')
plt.title('Pairplot of Numerical Variables with Streaming Platform as Hue')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='Subscription Type', y='Repeat Song Rate (%)', hue='Listening Time (Morning/Afternoon/Night)', palette='coolwarm')
plt.title('Subscription Type and Listening Time vs. Repeat Song Rate')
plt.show()
```
